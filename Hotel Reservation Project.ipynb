```python
# Hello, welcome to my Hotel Reservation System project created using Object-Oriented Programming! To start off, I imported
# the random module and created empty reservation lists for each type of room: Master, Penthouse, Double, and Single. 

import random 
reservation_master = []
reservation_penthouse = []
reservation_double = []
reservation_single = [] 

# Here I created the class Hotel(), with the attributes total cost, room rent cost, the customer's name, the customer's address,
# check in and check out date, and room number.

class Hotel:
    
    def __init__(self,total = '',roomrent = 0,name = '', address = '',checkin = '', checkout = '', rno = ''):
        
        print('******* WELCOME TO HOTEL SUPREME! *******')
        
        self.total = total
        self.roomrent = roomrent
        self.name = name
        self.address = address
        self.checkin = checkin
        self.checkout = checkout
        self.rno = rno
 
 # Customer_data(self) for the class Hotel takes user input from the customer for their full name, address, and 
  
 # check in/check out date, and also uses random.randint to randomly select a room number that will be printed on the screen

    def customer_data(self):
        self.name = input('\nWhat is your full name?:')
        self.address = input("\nWhat is your address?:")
        self.checkin = input("\nEnter the date you would like to check in:")
        self.checkout = input("\nEnter your check out date:")
        self.rno = random.randint(100,300)
        print("\nYou have been assigned Room Number:",self.rno,"\n")

 # Room_rent(self) prints out basic information about starting rates for the rooms Hotel Supreme has available to rent out 
 # and gives the user the option to input which room they would like to rent out and for how many nights would they like to stay
 # Customers can also calculate their total cost for their stay. 
        
    def room_rent(self):
     
        print("\nHere are our available suites and the rates they go for per night:")
        print("\nPenthouse: The most exclusive suite we offer, a rooftop paradise! Rates start at $2500/night.")
        print("\nMaster: A perfect suite for a honeymoon! Very romantic! Rates start at $2000/night.")
        print("\nDouble : A standard hotel room with two beds and a bathroom. Rates start at $700/night.")
        print("\nSingle: As simple as it gets, perfect for business/work trips. Rates start at $500/night.")
    
        rc = input("\nWhich room would you like to rent?")
        nights = int(input("\nFor how many nights did you stay in our establishment?"))

 # This section checks if there are any rooms left available to book for the time requested. It calculates the total cost 
 # based upon how many nights are selected for the room and which room is chosen, and adds it to the customer's bill. 
 #If room is available, a class instance  with customer name and room number are appended to the reservation list. 
 #If the room is not available, i.e the room is booked, then the program will inform you that the room is unavailable
        
        if (rc == 'Penthouse' and len(reservation_penthouse) < 2):
            
            print("\nIt's gonna be a party! Have fun!")
            self.roomrent =2500*nights
            reservation_penthouse.append([self.name,self.rno])
        else:
            print('Sorry, we are booked!')
        
        if (rc == 'Master' and len(reservation_master) < 4):
            
            print("\nYou and your spouse will love it! Enjoy your stay and Happy Honeymoon!")
            self.roomrent = 1750*nights
            reservation_master.append([self.name,self.rno])
        
        if (rc == 'Double' and len(reservation_double) < 5):
          
            print("\nSweet,simple,to the point. Enjoy your stay!")
            self.roomrent = 700*nights
            reservation_double.append([self.name,self.rno])
        
        if (rc == 'Single' and len(reservation_single) < 5):
        
            print("\nBusiness is Business. Enjoy the stay!")
            self.roomrent = 500*nights
            reservation_single.append([self.name,self.rno])
 
 # Bill() just takes customer information and displays it on a bill.

    def bill(self):
        print("******* SUPREME BILL *******")
        print('Customer Details:')
        print('Name:',self.name)
        print('Address:',self.address)
        print('Check in date:',self.checkin)
        print('Check out date:',self.checkout)
        print('Room Number:',self.rno)
        print('Room rent total:$',self.roomrent)
  
 # Roomcheck(self) enables a customer to check availability for each type of suite.

    def roomcheck(self):
        c = input("\nWhich room type would you like to check availability for?")
        
        if (c == 'Penthouse'):
            if len(reservation_penthouse) >= 2:
                print("\nWe are booked for Penthouse Suites. Try again another time.")
            else:
                print("\nPenthouse Available.")
                
        if (c == 'Master'):
            if len(reservation_master) >= 4:
                print("\nWe are booked for Master Suites. Try again another time.")
            else:
                print("\nMaster Available.")
                
        if (c == 'Double'):
            if len(reservation_double) >= 5:
                print("\nWe are booked for Double Suites. Try again another time.")
            else:
                print("\nDouble Available.")
                
        if (c == 'Single'):
            if len(reservation_penthouse) >= 5:
                print("\nWe are booked for Single Suites. Try again another time.")
            else:
                print("\nSingle Available.")
            
            
            
 # Menu() is a straight forward function that calls the hotel class and streamlines all of the functions I created into a 
 # simple and easy to use menu to choose if they want to enter their data, check room availability, calculate the total rent,
 # view their bill, or exit. 
    

def menu():
    hotel = Hotel()
    
    while(1):
        
        print('A. Enter Customer Data for Check-In')
        
        print('B. Check Room Availability')
        
        print('C. Calculate the rent total for your room')
        
        print('D. View your bill for your stay')
        
        print('E. Exit ')
        
        a = input("\nChoose your option:")
        
        if (a == 'A'):
            hotel.customer_data()
            
        if (a == 'B'):
            hotel.roomcheck()
            
        if (a == 'C'):
            hotel.room_rent()
            
        if (a == 'D'):
            hotel.bill()
        
        if (a == 'E'):
            break
    
        
```

```python
menu()
```
